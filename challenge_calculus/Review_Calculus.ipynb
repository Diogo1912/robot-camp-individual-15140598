# Bronze medal: Integral of a Polynomial

a) Student A carefully examine his data and proposed a model: 

The clicks per hour with respect to time can be described by function 

## $y=7.98 \times 10^{-5} \cdot x^5 - 1.80 \times 10^{-2} \cdot x^4 + 1.48 \cdot x^3 - 52.5 \cdot x^2 + 680 \cdot x$

where x is the past time in hour and y is the clicks. Can you plot the data and this function to have a look?


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# a) Load data and plot Student A's data with the proposed model

# Load Student A's data
data_A = pd.read_csv('data_student_A.csv')

# Define Student A's proposed model
def student_A_model(x):
    return (7.98e-5 * x**5 - 1.80e-2 * x**4 + 1.48 * x**3 - 52.5 * x**2 + 680 * x)

# Create x values for the model curve
x_model = np.linspace(0, 80, 1000)
y_model = student_A_model(x_model)

# Plot the data and model
plt.figure(figsize=(10, 6))
plt.plot(data_A['time (h)'], data_A[' clicks per hour'], 'bo', alpha=0.6, label='Student A Data')
plt.plot(x_model, y_model, 'r-', linewidth=2, label='Student A Model')
plt.xlabel('Time (hours)')
plt.ylabel('Clicks per hour')
plt.title('Student A: Clicks per Hour vs Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Model: y = 7.98×10⁻⁵x⁵ - 1.80×10⁻²x⁴ + 1.48x³ - 52.5x² + 680x")
```

# Silver medal: Find the derivative of a function using the product rule and the rule of composite function

a) Student B carefully examined his data and checked the literature. After that, he use this function to fit his data:

## $y=144000\cdot e^{-x/8}\left( -x - 8\right)+1152000$

where x is the past time in hour and y is the total viewing time (in the unit of minus). 

Can you find the deriviative of this function?

```python
# a) Find the derivative of Student B's function using product rule and chain rule

# Function: y = 144000 × e^(-x/8) × (-x - 8) + 1152000
# This can be written as: y = 144000 × e^(-x/8) × (-x - 8) + 1152000

# Let u = 144000 × e^(-x/8) and v = (-x - 8)
# Then the first part is u × v, and we use the product rule: (uv)' = u'v + uv'

# u = 144000 × e^(-x/8)
# u' = 144000 × e^(-x/8) × (-1/8) = -18000 × e^(-x/8)

# v = (-x - 8) 
# v' = -1

# Using product rule: (uv)' = u'v + uv'
# (uv)' = (-18000 × e^(-x/8)) × (-x - 8) + (144000 × e^(-x/8)) × (-1)
# (uv)' = 18000 × e^(-x/8) × (x + 8) - 144000 × e^(-x/8)
# (uv)' = e^(-x/8) × [18000(x + 8) - 144000]
# (uv)' = e^(-x/8) × [18000x + 144000 - 144000]
# (uv)' = e^(-x/8) × 18000x
# (uv)' = 18000x × e^(-x/8)

# The derivative of the constant 1152000 is 0

# Therefore: dy/dx = 18000x × e^(-x/8)

print("Student B's function: y = 144000 × e^(-x/8) × (-x - 8) + 1152000")
print("Using product rule and chain rule:")
print("dy/dx = 18000x × e^(-x/8)")

def student_B_derivative(x):
    return 18000 * x * np.exp(-x/8)
```

b) Following their previous discussion, assuming the correctness of the function found by Student B, they should be able to find the function that the number of clicks per hour will obey. Can you help them find this function?

You can plot the data of student A and this function to check that.

```python
def DerivativeExp(x):
    # This is the derivative we found: dy/dx = 18000x × e^(-x/8)
    # This represents clicks per hour as a function of time
    y = 18000 * x * np.exp(-x/8)
    return y

# b) Plot this function against Student A's data to check
x_model = np.linspace(0, 80, 1000)
y_model = [DerivativeExp(x) for x in x_model]

plt.figure(figsize=(10, 6))
plt.plot(data_A['time (h)'], data_A[' clicks per hour'], 'bo', alpha=0.6, label='Student A Data')
plt.plot(x_model, y_model, 'g-', linewidth=2, label='Student B Derivative (Exponential)')
plt.xlabel('Time (hours)')
plt.ylabel('Clicks per hour')
plt.title('Comparison: Student A Data vs Student B Derivative Model')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Student B's derivative function: f'(x) = 18000x × e^(-x/8)")
print("This represents the clicks per hour derived from the total viewing time function")
```

# Gold medal: Find the derivative of a function using the quotient rule and conducting optimization

a) Student C found their work very interesting and got involved. In some other paper, Student C had read that the Sigmoid function was very useful for this type of data and he decided to create such a model. After a regression to get the constants, his function was::

## $f(x) = \left(\frac{1.65 \times 10^6} {1 + \exp{\left(-\frac{x}{8} + 1\right)}} \right)-5\times 10^5$

Can you help Student C find the function that demonstrates clicks per hour as a function of time?

```python
def DerivativeSig(x):
    # a) Find the derivative of Student C's sigmoid function using quotient rule and chain rule
    
    # Function: f(x) = (1.65×10⁶)/(1 + exp(-x/8 + 1)) - 5×10⁵
    
    # Let's break this down:
    # f(x) = A/(1 + exp(B)) - C
    # where A = 1.65×10⁶, B = -x/8 + 1, C = 5×10⁵
    
    # The derivative of the constant C is 0
    # For the first part, we use the quotient rule and chain rule
    
    # Let u = A = 1.65×10⁶ (constant), v = 1 + exp(-x/8 + 1)
    # (u/v)' = (u'v - uv')/v²
    # u' = 0, v' = exp(-x/8 + 1) × (-1/8) = -exp(-x/8 + 1)/8
    
    # (u/v)' = (0 × v - A × (-exp(-x/8 + 1)/8))/v²
    # (u/v)' = (A × exp(-x/8 + 1)/8)/(1 + exp(-x/8 + 1))²
    
    A = 1.65e6
    exp_term = np.exp(-x/8 + 1)
    denominator = 1 + exp_term
    
    # Derivative: f'(x) = (A × exp(-x/8 + 1)/8) / (1 + exp(-x/8 + 1))²
    y = (A * exp_term / 8) / (denominator**2)
    
    return y

# Test the function
x_test = np.linspace(0, 80, 1000)
y_test = [DerivativeSig(x) for x in x_test]

plt.figure(figsize=(10, 6))
plt.plot(data_A['time (h)'], data_A[' clicks per hour'], 'bo', alpha=0.6, label='Student A Data')
plt.plot(x_test, y_test, 'm-', linewidth=2, label='Student C Derivative (Sigmoid)')
plt.xlabel('Time (hours)')
plt.ylabel('Clicks per hour')
plt.title('Student C: Sigmoid Model Derivative vs Student A Data')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Student C's sigmoid derivative:")
print("f'(x) = (1.65×10⁶ × exp(-x/8 + 1)/8) / (1 + exp(-x/8 + 1))²")
```

```python
# b) Compare all three models to determine the best one

# Plot all models together for comparison
plt.figure(figsize=(12, 8))

# Student A's original polynomial model  
x_model = np.linspace(0, 80, 1000)
y_poly = [student_A_model(x) for x in x_model]
plt.plot(x_model, y_poly, 'r-', linewidth=2, label='Student A: Polynomial Model')

# Student B's derivative (exponential)
y_exp = [DerivativeExp(x) for x in x_model]
plt.plot(x_model, y_exp, 'g-', linewidth=2, label='Student B: Exponential Derivative')

# Student C's derivative (sigmoid)
y_sig = [DerivativeSig(x) for x in x_model]
plt.plot(x_model, y_sig, 'm-', linewidth=2, label='Student C: Sigmoid Derivative')

# Original data
plt.plot(data_A['time (h)'], data_A[' clicks per hour'], 'bo', alpha=0.7, label='Student A Data', markersize=4)

plt.xlabel('Time (hours)')
plt.ylabel('Clicks per hour')
plt.title('Comparison of All Three Models')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Calculate some basic goodness-of-fit metrics
from scipy.stats import pearsonr

# Get data points for comparison
data_points = data_A[' clicks per hour'].values
time_points = data_A['time (h)'].values

# Calculate predictions for each model at data points
poly_pred = [student_A_model(t) for t in time_points]
exp_pred = [DerivativeExp(t) for t in time_points]
sig_pred = [DerivativeSig(t) for t in time_points]

# Calculate correlation coefficients
corr_poly, _ = pearsonr(data_points, poly_pred)
corr_exp, _ = pearsonr(data_points, exp_pred)
corr_sig, _ = pearsonr(data_points, sig_pred)

print("Model Comparison:")
print(f"Polynomial Model Correlation: {corr_poly:.4f}")
print(f"Exponential Model Correlation: {corr_exp:.4f}")
print(f"Sigmoid Model Correlation: {corr_sig:.4f}")

# Calculate Mean Squared Error
mse_poly = np.mean([(d - p)**2 for d, p in zip(data_points, poly_pred)])
mse_exp = np.mean([(d - p)**2 for d, p in zip(data_points, exp_pred)])
mse_sig = np.mean([(d - p)**2 for d, p in zip(data_points, sig_pred)])

print(f"\nMean Squared Error:")
print(f"Polynomial Model MSE: {mse_poly:.2f}")
print(f"Exponential Model MSE: {mse_exp:.2f}")
print(f"Sigmoid Model MSE: {mse_sig:.2f}")
```

c) Use the best model to calculate: at which hour does clicks per hour reach its maximum value? (rounded to two decimal places)

```python
# c) Find the time when clicks per hour reach maximum value

from scipy.optimize import minimize_scalar

def find_max_clicks(BestModel):
    # To find the maximum, we can either:
    # 1. Take the derivative and find where it equals zero
    # 2. Use numerical optimization
    
    # For the polynomial model, let's find the derivative and solve for critical points
    # f(x) = 7.98×10⁻⁵x⁵ - 1.80×10⁻²x⁴ + 1.48x³ - 52.5x² + 680x
    # f'(x) = 5×7.98×10⁻⁵x⁴ - 4×1.80×10⁻²x³ + 3×1.48x² - 2×52.5x + 680
    # f'(x) = 3.99×10⁻⁴x⁴ - 7.2×10⁻²x³ + 4.44x² - 105x + 680
    
    def derivative(x):
        return (3.99e-4 * x**4 - 7.2e-2 * x**3 + 4.44 * x**2 - 105 * x + 680)
    
    # Use numerical method to find maximum in reasonable range [0, 80]
    # We minimize the negative of the function to find maximum
    result = minimize_scalar(lambda x: -BestModel(x), bounds=(0, 80), method='bounded')
    
    return result.x

# Print the maximum clicks per hour
max_time = find_max_clicks(BestModel)
print("The maximum clicks per hour occur at {:.2f}.".format(max_time))
```

d) This YouTuber happened to get a double-coupon. This special double-coupon can doubles his profits from all viewings for 5 hours and he can choose the exact time to start. Based on the model we found, can you suggest which hour he should start using this coupon? The time is counted as 0 when his video is published. (rounded to two decimal places)

```python
# d) Find the best time to start a 5-hour double coupon

from scipy.optimize import minimize_scalar
from scipy.integrate import quad

def find_best_coupon_time(BestModel):
    # We want to maximize the total profit over a 5-hour period
    # This means we want to maximize the integral of the function over any 5-hour window
    
    def total_profit_in_window(start_time):
        # Calculate the definite integral from start_time to start_time + 5
        # This represents the total clicks (and thus profit) in the 5-hour window
        
        if start_time < 0 or start_time + 5 > 80:
            return 0  # Invalid window
            
        integral_result, _ = quad(BestModel, start_time, start_time + 5)
        return integral_result
    
    # Find the start time that maximizes the 5-hour integral
    # We minimize the negative to find the maximum
    result = minimize_scalar(lambda t: -total_profit_in_window(t), 
                           bounds=(0, 75), method='bounded')  # 75 so that window ends at 80
    
    return result.x

# Finding best 5-hour period to activate the double coupon
best_start_time = find_best_coupon_time(BestModel)
print("The best time to start the 5-hour double coupon is at hour {:.2f}".format(best_start_time))

# Verify by showing the profit in this window
start = best_start_time
end = start + 5
total_clicks_in_window, _ = quad(BestModel, start, end)
print(f"Total clicks in optimal 5-hour window ({start:.2f} to {end:.2f}): {total_clicks_in_window:.0f}")
```
